## Krok 1: üìã Jak vypad√° soubor CSV?

CSV (Comma-Separated Values) je textov√Ω form√°t ‚Äì **prvn√≠ ≈ô√°dek** obsahuje n√°zvy sloupc≈Ø, **dal≈°√≠ ≈ô√°dky** jsou data. Hodnoty jsou oddƒõlen√© ƒç√°rkou.

````python
# üìå D≈Øle≈æit√©
with open("documents/studenti.csv", encoding="utf-8") as soubor:
    print(soubor.read())
````



Soubor otev≈ôe≈° pomoc√≠ `open()`. D≈Øle≈æit√© je **`encoding="utf-8"`** ‚Äì bez nƒõj by se ƒçesk√© znaky (ƒõ, ≈°, ƒç‚Ä¶) mohly zobrazit ≈°patnƒõ.

**`with open(...) as soubor:`** znamen√°, ≈æe soubor se po skonƒçen√≠ bloku s√°m zav≈ôe ‚Äì nemus√≠≈° na to myslet.

## Krok 2: üìñ ≈ò√°dky jako ‚Äûslovn√≠ky‚Äú ‚Äì co to znamen√°?

Abychom z ka≈æd√©ho ≈ô√°dku mohli br√°t hodnoty podle **n√°zvu sloupce** (t≈ôeba ‚Äûchci jm√©no‚Äú), pou≈æije≈° dal≈°√≠ hodnotu v Pythonu, **slovn√≠k**.

**Slovn√≠k** je struktura, kde m√°≈° dvojice **kl√≠ƒç ‚Üí hodnota**:
- Kl√≠ƒç je t≈ôeba jm√©no (t≈ôeba n√°zev sloupce),
- hodnota je √∫daj (jm√©no studenta, poƒçet bod≈Ø‚Ä¶).

K hodnotƒõ se dostane≈° z√°pisem `slovn√≠k['jmeno']`.

*Pokud slovn√≠k nezn√°≈°, nevad√≠ ‚Äì v dal≈°√≠ bu≈àce je jeden ≈ô√°dek ‚Äûjako by z CSV‚Äú.*

````python
# Jeden ‚Äû≈ô√°dek‚Äú jako slovn√≠k ‚Äì kl√≠ƒçe = n√°zvy sloupc≈Ø, hodnoty = √∫daje
ukazkovy_radek = {"jmeno": "Anna Nov√°kov√°", "trida": "1.A", "body": "82"}
````

````python

print(ukazkovy_radek)
````

````python
print(ukazkovy_radek["jmeno"])
print(ukazkovy_radek["body"])
````

**`csv.DictReader`** dƒõl√° to sam√© automaticky: **prvn√≠ ≈ô√°dek** CSV (hlaviƒçka) bere jako n√°zvy sloupc≈Ø neboli kl√≠ƒçe, **ka≈æd√Ω dal≈°√≠ ≈ô√°dek** p≈ôevede na takov√Ω slovn√≠k. M≈Ø≈æe≈° pak ps√°t `radek['jmeno']` m√≠sto ‚Äût≈ôet√≠ sloupec‚Äú ‚Äì je to p≈ôehlednƒõj≈°√≠ a m√©nƒõ n√°chyln√© k chyb√°m.

````python
import csv

with open('documents/studenti.csv', encoding='utf-8') as soubor:
    obsah_csv = csv.DictReader(soubor)
````

````python
print(obsah_csv)
````

`csv.DictReader` n√°m vrac√≠ speci√°ln√≠ objekt, kter√Ω m≈Ø≈æeme proch√°zet jen jednou. 
Proto ho p≈ôevedeme na seznam pomoc√≠ `list()`, abychom s n√≠m mohli pracovat v√≠cekr√°t.

````python
with open('documents/studenti.csv', 'r', encoding='utf-8') as soubor:
    obsah_csv = csv.DictReader(soubor)
    nactene_radky = list(obsah_csv)
````

````python
print(nactene_radky)
````

## Krok 3: üîÑ Kontrola v≈°ech ≈ô√°dk≈Ø a p≈ô√≠stup k hodnot√°m

Teƒè m√°≈° kompletn√≠ data, naƒçten√° v Pythonu, ale jak si je postupnƒõ proch√°zet?

````python
for radek in nactene_radky:
    print(radek)
````


Pokud **cyklus `for`** je≈°tƒõ nezn√°≈°: je to zp≈Øsob, jak ≈ô√≠ct Pythonu *‚Äûudƒõlej to sam√© pro ka≈ædou jednu vƒõc v seznamu (nebo v ≈ô√°dc√≠ch souboru)‚Äú*.

Z√°pis `for radek in nactene_radky:` znamen√°: *‚ÄûVezmi z `nactene_radky` jeden ≈ô√°dek po druh√©m, ulo≈æ ho do promƒõnn√© `radek`, a pro ka≈æd√Ω takov√Ω ≈ô√°dek vykonej odsazen√Ω blok k√≥du.‚Äú*

V cyklu pak z ka≈æd√©ho `radek` bere≈° hodnoty podle **n√°zvu sloupce**: `radek['jmeno']`, `radek['trida']`, `radek['body']`.

````python
for radek in nactene_radky:
    print(radek['body'])
````

## Krok 4: üî¢ Pravda nebo ne?

Aktu√°lnƒõ m√°≈° p≈ô√≠stup k bod≈Øm pro jednotliv√© studenty:
```bash
82
68
91
75
59
88
```

Jak teƒè ale ovƒõ≈ô√≠≈°, kter√Ω student m√° v√≠ce bod≈Ø ne≈æ je limit, tedy **85 bod≈Ø**?



### Podm√≠nkov√Ω z√°pis


````python
hodnoceni = 74
````

````python
if hodnoceni > 85:
    print("√öspƒõ≈°n√Ω")
else:
    print("Ne√∫spƒõ≈°n√Ω")
````

Z√°pis **`if hodnoceni >= 75:`** znamen√°: *‚ÄûKdy≈æ je hodnota v `hodnoceni` vƒõt≈°√≠ nebo rovna 75, vykonej odsazen√Ω blok pod t√≠mto `if`.

Jinak ho p≈ôeskoƒç a proveƒè automaticky z√°pis pod `else`.


````python
for radek in nactene_radky:
    if radek['body'] > 85:
        print("√öspƒõ≈°n√Ω")
    else:
        print("Ne√∫spƒõ≈°n√Ω")
````

Tady ale porovn√°v√°m dvƒõ r≈Øzn√© hodnoty. Podle Python `str` a `int`, co≈æ zp≈Øsob√≠ chybu.

````python
print(type(radek['body']))
````

````python
print(type(85))
````


Abychom mohli porovn√°vat body (‚Äûje to 75 nebo v√≠c?‚Äú), mus√≠me p≈ôev√©st string na cel√© ƒç√≠slo: **`int(radek['body'])`**.

````python
print(type(int(radek['body'])))
````

Po drobn√© √∫pravƒõ:

````python
for radek in nactene_radky:
    if int(radek['body']) > 85:
        print("√öspƒõ≈°n√Ω,", radek['jmeno'])
    else:
        print("Ne√∫spƒõ≈°n√Ω")
````

---

## Krok 5: ‚úÖ Cel√Ω program ‚Äì v√Ωpis jen √∫spƒõ≈°n√Ωch

Spoj√≠me otev≈ôen√≠ souboru, `DictReader`, cyklus, podm√≠nkov√Ω z√°pis a p≈ôevod `int(radek['body'])` dohromady.

````python
import csv

with open('documents/studenti.csv', encoding='utf-8') as soubor:
    obsah_csv = csv.DictReader(soubor)
    nactene_radky = list(obsah_csv)

print("Studenti s 85 a v√≠ce body:\n")

for radek in nactene_radky:
    if int(radek['body']) > 85:
        print("√öspƒõ≈°n√Ω,", radek['jmeno'])
    else:
        print("Ne√∫spƒõ≈°n√Ω")  
````

## üìù Co u≈æ po tomto cviƒçen√≠ zn√°≈°?

| Koncept | P≈ô√≠klad |
|---------|--------|
| Otev≈ôen√≠ souboru | `with open('soubor.csv', encoding='utf-8') as soubor:` |
| `csv.DictReader` | ≈ò√°dky jako slovn√≠ky, kl√≠ƒçe = n√°zvy sloupc≈Ø |
| P≈ô√≠stup ke sloupci | `radek['jmeno']`, `radek['body']` |
| P≈ôevod na ƒç√≠slo | `int(radek['body'])` |
| Podm√≠nka | `if body >= 75:` |
| For smyƒçku | `for radek in nactene_radky` |
